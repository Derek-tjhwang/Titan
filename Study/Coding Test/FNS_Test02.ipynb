```python
import csv
import operator

f = open('data_1.csv', 'r', encoding='utf-8')
lines = csv.reader(f)
f1 = True
partners = {}

for line in lines:
    if f1 == True:
        f1 = False
        pass
    else:
        if line[2] in partners.keys():
            partners[line[2]] = partners[line[2]] + 1
        else:
            partners[line[2]] = 1
        #print(line[2])

sorted_partners = sorted(partners.items(), key=operator.itemgetter(1))
print(sorted_partners)
f.close()
```
