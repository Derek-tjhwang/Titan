```python
import csv
import datetime
import time
import math
from matplotlib import pyplot as plt


def normal_pdf(x, mu, sigma):
    sqrt_two_pi = math.sqrt(2*math.pi)
    return (math.exp(-(x-mu) ** 2 / 2 / sigma ** 2) / (sqrt_two_pi * sigma))


f = open('data_1.csv', 'r', encoding='utf-8')
lines = csv.reader(f)
partnerst1 = {'462':[],'26704':[],'31340':[],'49423':[],'46922':[]}
partnerst2 = {'462':[],'26704':[],'31340':[],'49423':[],'46922':[]}
partnerstre = {'462':[],'26704':[],'31340':[],'49423':[],'46922':[]}
partnerstmu = {'462':0,'26704':0,'31340':0,'49423':0,'46922':0}
partnerstsig = {'462':0,'26704':0,'31340':0,'49423':0,'46922':0}                                                                  

for line in lines:
    if line[2] in partnerst1.keys():
        partnerst1[line[2]].append(datetime.datetime.strptime(line[0],"%Y-%m-%dT%H:%M:%S"))
        partnerst2[line[2]].append(datetime.datetime.strptime(line[1],"%Y-%m-%dT%H:%M:%S"))


for k in partnerstre.keys():
    print(k)
    #print(len(partnerst1[k]))
    for i in range(0, len(partnerst1[k])):
        result = partnerst1[k][i] - partnerst2[k][i]
        partnerstre[k].append(result.total_seconds() / 3600)
        #print(partnerstre[k][i])
        # 시간 SUM에 더하기
        partnerstmu[k] = partnerstmu[k] + float(result.total_seconds() / 3600)
                                                                  
    # 시간 합에 총 아이템의 갯수로 나누기
    partnerstmu[k] = partnerstmu[k] / len(partnerst1[k])
    print(partnerstmu[k])
    
# mu 값을 토대로 표준 편차를 구함
for k in partnerstsig.keys():
    for i in range(0, len(partnerstre[k])):
        partnerstsig[k] = partnerstsig[k] + (partnerstre[k][i] - partnerstmu[k]) ** 2
    partnerstsig[k] = math.sqrt(partnerstsig[k] / len(partnerstre[k]))
    print(partnerstsig[k])
    

# 표준 편차를 토대로 정규 분포를 그린다.
plt.plot(partnerstre['462'], [normal_pdf(x, partnerstmu['462'], partnerstsig['462']) for x in partnerstre['462']], '-', label='462')
plt.plot(partnerstre['26704'], [normal_pdf(x, partnerstmu['26704'], partnerstsig['26704']) for x in partnerstre['26704']], '--', label='26704')
plt.plot(partnerstre['31340'], [normal_pdf(x, partnerstmu['31340'], partnerstsig['31340']) for x in partnerstre['31340']], ':', label='31340')
plt.plot(partnerstre['49423'], [normal_pdf(x, partnerstmu['49423'], partnerstsig['49423']) for x in partnerstre['49423']], '-.', label='49423')
#plt.plot(partnerstre['46922'], [normal_pdf(x, partnerstmu['46922'], partnerstsig['46922']) for x in partnerstre['46922']], '-', label='46922')
plt.legend()
plt.title("Normal Pdfs")
plt.show()

f.close()
```
