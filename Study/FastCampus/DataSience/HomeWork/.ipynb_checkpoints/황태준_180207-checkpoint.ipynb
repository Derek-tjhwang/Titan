## Problem 1

```python
import random
import numpy as np

# n = 10의 numpy 벡터 생성
def create_li(x):
    random.seed(x)
    return np.array([random.randrange(1,11) for _ in range(10)])

# 벡터의 길이 반환
def get_vec_len(li_x):
    return np.sqrt(sum(np.square(li_x)))

li_a = create_li(43)
li_b = create_li(34)

vl_a = get_vec_len(li_a)
vl_b = get_vec_len(li_b)

nv_a = li_a/vl_a
nv_b = li_b/vl_b

print np.sqrt(sum(np.square(nv_a-nv_b)))
```

## Problem 5

```python
import random
import numpy as np

def get_gradient(v):
    return [4*(v_i**3) for v_i in v]

def step(v, gradient, step_size):
    return [v_i + step_size * direction_i 
            for v_i, direction_i in zip(v, gradient)]

def distance(v, w):
    return np.sqrt(sum([(v_i - w_i)**2 for v_i, w_i in zip(v, w)]))

v = [random.randint(-10, 10) for _ in range(3)]
alpha = -0.01

while True:
    gradient = get_gradient(v)
    next_v = step(v, gradient, alpha)
    print next_v
    if distance(next_v, v) < 0.000001:
        break
    v = next_v
    
print "극소값은 %f %f %f 입니다. " %(v[0], v[1], v[2])
```

## Problem 8

```python
relation_li = [{0,1},{0,2},{1,2},{1,3},{2,3},
               {3,4},{4,5},{5,6},{5,7},{6,8},{7,8},{8,9}]

def check_relation(x):
    re_li = []
    flag = 0
    for p, v in relation_li:
        if p == x:
            re_li.append(v)
            flag = 1
        else:
            if flag:
                break
    return re_li
    
i = int(raw_input())
check_relation(i)
```
